```python
%pip install qiskit==1.2.4
%pip install qiskit-aer==0.15.1
%pip install pylatexenc==2.10

from qiskit import QuantumCircuit
from qiskit.converters import circuit_to_gate
from qiskit.visualization import array_to_latex
from qiskit.quantum_info import Operator
from qiskit import transpile
from qiskit.providers.basic_provider import BasicSimulator
from qiskit.providers.fake_provider import GenericBackendV2
from qiskit.visualization import plot_histogram
from qiskit.result import marginal_counts
```

```python
# Teleportation
# Create a circuit with 3 qubits (first parameter) and 3 classical bits (second parameter)
teleport = QuantumCircuit(3,3)

# qubit 0 is the state to be teleported from Alice
# qubits 1 and 2 are the shared entangled state
# Alice has qubit 1, Bob has qubit 2

# Add gates here to entangle qubits 1 and 2
# H on qubit 1, then CNOT with qubit 1 as control and qubit 2 as target
# This creates the Bell state (1/sqrt(2))(|00> + |11>) on qubits 1,2
teleport.h(1)
teleport.cx(1, 2)

# Add gates here to entangle qubit 0 with the others
# Alice applies CNOT with qubit 0 (the state to teleport) as control, qubit 1 as target
# Then applies H to qubit 0
teleport.cx(0, 1)
teleport.h(0)

# Measure qubits 0 and 1 and put the results on classical bits 0 and 1.
teleport.measure([0,1],[0,1])

# Fix up the final state of qubit 2, depending on the measurement results.
# Here we are using a feature of Qiskit that we have not seen before.
# The second classical bit (bit 1) controls X on qubit 2
# The first classical bit (bit 0) controls Z on qubit 2
with teleport.if_test((1, 1)):
    teleport.x(2)
with teleport.if_test((0, 1)):
    teleport.z(2)

# Sometimes it's useful to insert a "barrier" to make the diagram clearer.
# A barrier also has an effect on how the circuit is compiled, but we won't worry about that.
teleport.barrier()

# Measure qubit 2 (first parameter) and put the result on classical bit 2 (second parameter)
teleport.measure([2],[2])

# Check the circuit
teleport.draw("mpl")
```

```python
# Use a circuit to put qubit 0 into the state that we want to teleport,
# and compose it into the main circuit.

prep = QuantumCircuit(1)
# Add a gate here to put qubit 0 into some state other than |0>
# Here we use H to put qubit 0 into |+> = (1/sqrt(2))(|0> + |1>)
# This is a non-trivial superposition state to verify teleportation works
prep.h(0)

# When we compose circuits, the second parameter is the list of qubits (wires) that are being connected
teleport = teleport.compose(prep, [0], front=True)
teleport.draw("mpl")
```

```python
# Now simulate and plot the results.
# In general this won't tell us that we have exactly the right state,
# just that we get the right measurement statistics.
# We are interested in the result on classical bit 2, which is the most significant bit (leftmost).
# The simulator adds some noise, so we get some non-zero probabilities for parts of bit 2 that shouldn't be there.
# To see results without noise, add the parameter noise_info=False to GenericBackendV2
# We need to use GenericBackendV2 instead of BasicSimulator, and set control_flow=True, so that we can use if_test.

backend = GenericBackendV2(num_qubits=3, control_flow=True)
compiled = transpile(teleport, backend)
job_sim = backend.run(compiled, shots=1024)
result_sim = job_sim.result()
counts = result_sim.get_counts(compiled)
mcounts = marginal_counts(counts,indices=[2]) # We only want to see the results for bit 2
plot_histogram(mcounts)
```

```python
# Teleportation has an interesting feature which is not usually
# included in the description.

# If the qubit being teleported is entangled with other qubits, then after
# teleportation it is still entangled in the same way.

# Try this out!
# You will need to introduce another qubit and entangle it with qubit 0.
# After teleporting, work out how to use measurements to convince yourself that qubit 2 is entangled with the extra qubit.
# You can do this in a similar way to the exercise in Lab1 about Bell states.

# SOLUTION:
# We extend to 4 qubits and 4 classical bits.
# qubit 3 is the extra qubit that we entangle with qubit 0 (the state being teleported).
# We create a Bell pair between qubit 3 and qubit 0 BEFORE teleportation.
# After teleportation, qubit 2 should be entangled with qubit 3 in the same way.
# To verify: measure qubit 3 and qubit 2 in the Bell basis.
# They should be perfectly correlated (both 0 or both 1) showing entanglement was preserved.

teleport_entangled = QuantumCircuit(4, 4)

# Step 1: Entangle qubit 3 (extra) with qubit 0 (state to teleport)
# This creates Bell state (1/sqrt(2))(|00> + |11>) on qubits 3,0
teleport_entangled.h(3)
teleport_entangled.cx(3, 0)

# Step 2: Entangle qubits 1 and 2 (shared Bell pair between Alice and Bob)
teleport_entangled.h(1)
teleport_entangled.cx(1, 2)

# Step 3: Alice applies CNOT and H to qubits 0 and 1
teleport_entangled.cx(0, 1)
teleport_entangled.h(0)

# Step 4: Alice measures qubits 0 and 1
teleport_entangled.measure([0, 1], [0, 1])

# Step 5: Bob applies corrections to qubit 2 based on Alice's measurement
with teleport_entangled.if_test((1, 1)):
    teleport_entangled.x(2)
with teleport_entangled.if_test((0, 1)):
    teleport_entangled.z(2)

teleport_entangled.barrier()

# Step 6: Verify entanglement between qubit 2 (Bob) and qubit 3 (extra)
# Decode the Bell basis by applying CNOT then H, then measure
# If qubit 2 and qubit 3 are in the same Bell state as qubits 3 and 0 were,
# after the decode they should both measure 0.
teleport_entangled.cx(3, 2)
teleport_entangled.h(3)
teleport_entangled.measure([2, 3], [2, 3])

teleport_entangled.draw("mpl")
```

```python
# Simulate the entanglement-preserving teleportation circuit
# After the Bell basis decode, qubits 2 and 3 should both measure 0
# confirming that entanglement was preserved through teleportation.

backend = GenericBackendV2(num_qubits=4, control_flow=True)
compiled = transpile(teleport_entangled, backend)
job_sim = backend.run(compiled, shots=1024)
result_sim = job_sim.result()
counts = result_sim.get_counts(compiled)
# Marginalise to only bits 2 and 3 (Bob and extra qubit)
mcounts = marginal_counts(counts, indices=[2, 3])
plot_histogram(mcounts)
```
